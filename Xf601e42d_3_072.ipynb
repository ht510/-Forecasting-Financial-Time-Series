```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python


# Load the recession probabilities from 'RecessionIndicator.csv'
recession_probs = pd.read_csv('RecessionIndicator.csv', index_col=0, parse_dates=[0])
recession_probs.index.name = 'Date'  # Name the index as 'Date'
```

```python

# Shift the probabilities by 1 month to create a lagged version
recession_probs_lagged = recession_probs.shift(1)
```

```python

# Load the financial data from '2023-09-TF.csv'
financial_data = pd.read_csv('2023-09-TF.csv', index_col='sasdate', parse_dates=True)
```

```python

# Identify the columns that belong to Group 8 (Financial Variables)
# Replace with the actual column names from your dataset
group_8_columns = ['S&P 500', 'S&P: indust', 'S&P div yield', 'S&P PE ratio', 'FEDFUNDS', 'CP3Mx', 'TB3MS', 'TB6MS', 'GS1', 'GS5', 'GS10', 'AAA', 'BAA']  # Example columns
financial_data_group_8 = financial_data[group_8_columns]
```

```python

# Merge with recession probabilities
data_merged = financial_data_group_8.join(recession_probs_lagged)
```

```python

# Regression Analysis
for column in financial_data_group_8.columns:
    Y = data_merged[column]
    X = data_merged['SVC']  # Assuming SVC is the chosen model in RecessionIndicator.csv
    X = sm.add_constant(X)  # Adds a constant term to the predictor
    model = sm.OLS(Y, X, missing='drop').fit()  # OLS regression, dropping missing values
    print(f'Regression Results for {column}')
    print(model.summary())
    print('\n')

    # Optional: Plotting the results
    plt.figure(figsize=(10, 6))
    plt.scatter(X.iloc[:, 1], Y, alpha=0.3)  # Plotting the data points
    plt.plot(X.iloc[:, 1], model.predict(X), color='red')  # Regression line
    plt.title(f'Regression of {column} on Recession Probability')
    plt.xlabel('Recession Probability (lagged)')
    plt.ylabel(column)
    plt.show()
```
